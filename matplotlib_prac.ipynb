```python
import numpy as np
x = np.linspace(0, 15, 30)
y = x**2
plt.plot(x,y)
```

```python
plt.plot(x,y, "go-") # green(g),circles(o),line(-)
```

```python
plt.plot(x,y, "rd-", linewidth=.5, markersize=5) # **kwargs
```

```python
y1 = x**2.0
y2 = x**1.5
plt.plot(x, y1, "go-", label="GreenLine")
plt.plot(x, y2, "bs", label="BlueSquares")
plt.xlabel("X")
plt.ylabel("$Y$") # LaTex
plt.axis([0, 10.5, 0, 150 ]) # [xmin, xmax, ymin, ymax]
plt.legend(loc="upper left")
# plt.savefig("myfirstplot.pdf") to save file to the working directory
```

```python
x = np.logspace(-1,1,30)
y1 = x**2.0
y2 = x**3.5
plt.loglog(x, y1, "go-")
plt.loglog(x, y2, "bs-")
```

```python
x = np.random.normal(size=1000)
plt.hist(x)
```

```python
plt.hist(x, bins=np.linspace(-5,5,25))
```

```python
x = np.random.gamma(2,3,100000)
plt.hist(x, bins=30)
```

```python
plt.figure()
plt.subplot(221)
plt.hist(x, bins=30)
plt.subplot(222)
plt.hist(x, bins=30, cumulative=30)
plt.subplot(223)
plt.hist(x, bins=30, cumulative=30, histtype="step")
```
